This is a project that is going to utilize the use of Multivariate Regression to come up with investment decisions on the Property Market.
In reality we can speculate on many different aspects that can influence our decisions when purchasing properties, this project is going to take full advantage of making data driven decisions.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
```

Importing the necessary data for our regress model analysis.

```python
data = pd.read_excel(r"C:\Users\obien\OneDrive\Desktop\Projects\Regression_Analysis\Housing.xlsx")
data
```

Assigning independent and dependant variables for the model.

```python
x = data[['House Size (sq.ft.)', 'Number of Rooms', 'Year of Construction']]
y = data['House Price']
```

Running and initializing the model using OLS(Ordinary Least Squares) method to get the model.

```python
x1 = sm.add_constant(x)
reg = sm.OLS(y, x1).fit()
reg.summary()
```
